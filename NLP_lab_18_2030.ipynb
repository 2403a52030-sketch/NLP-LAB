```python
# Load pretrained summarization model and tokenizer
from transformers import BartForConditionalGeneration, BartTokenizer

model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

# Define a custom summarizer function that uses the loaded model and tokenizer
def summarizer(text, max_length, min_length, do_sample):
    inputs = tokenizer([text], max_length=1024, return_tensors='pt', truncation=True)
    summary_ids = model.generate(
        inputs['input_ids'],
        num_beams=4,
        max_length=max_length,
        min_length=min_length,
        early_stopping=True,
        do_sample=do_sample
    )
    return [{'summary_text': tokenizer.decode(s, skip_special_tokens=True, clean_up_tokenization_spaces=False) for s in summary_ids}]

print("Model and custom summarizer function loaded successfully!")
```

```python
# Sample dataset (you can replace with your own 5–8 texts)

texts = [
    """Artificial Intelligence is transforming industries worldwide. It enables machines to learn from data and make decisions...
    AI is used in healthcare, finance, and transportation. However, ethical concerns remain regarding bias and privacy.""",

    """Climate change is one of the biggest challenges facing humanity today. Rising temperatures are causing melting ice caps...
    Governments and organizations are working together to reduce carbon emissions and promote sustainability.""",

    """In recent years, online education has gained popularity. Platforms like Coursera and edX provide access to courses...
    However, lack of interaction and motivation are major challenges in online learning.""",

    """The software development lifecycle includes planning, designing, coding, testing, and deployment...
    Agile methodologies have improved flexibility and collaboration in teams.""",

    """A group of friends went on a hiking trip in the mountains. They enjoyed the scenic views and fresh air...
    The journey taught them teamwork and perseverance."""
]
```

```python
# Generate summaries
for i, text in enumerate(texts):
    summary = summarizer(text, max_length=100, min_length=30, do_sample=False)

    print(f"\n--- TEXT {i+1} ---")
    print("Original:", text[:150], "...")

    print("\nSummary:", summary[0]['summary_text'])
```

```python
# Experiment with different parameters

text = texts[0]

print("\n--- SHORT SUMMARY ---")
short_summary = summarizer(text, max_length=50, min_length=20, do_sample=False)
print(short_summary[0]['summary_text'])

print("\n--- MEDIUM SUMMARY ---")
medium_summary = summarizer(text, max_length=100, min_length=40, do_sample=False)
print(medium_summary[0]['summary_text'])

print("\n--- LONG SUMMARY ---")
long_summary = summarizer(text, max_length=150, min_length=60, do_sample=False)
print(long_summary[0]['summary_text'])
```

```python
# Simple comparison display

types = ["News", "Environment", "Education", "Technical", "Narrative"]

for i in range(len(texts)):
    summary = summarizer(texts[i], max_length=80, min_length=30, do_sample=False)

    print(f"\n--- {types[i]} TEXT ---")
    print("Summary:", summary[0]['summary_text'])
```

```python
long_text = """Artificial Intelligence (AI) has rapidly evolved over the past decade, impacting numerous sectors including healthcare, education, and finance...
It has enabled automation, predictive analytics, and improved decision-making processes across industries. However, concerns about job displacement,
data privacy, and ethical implications continue to grow. Governments and organizations are working to establish guidelines and regulations to ensure
responsible AI development. As AI continues to advance, it is expected to play a critical role in shaping the future of humanity.""" * 3

# Summarize long text
summary = summarizer(long_text, max_length=120, min_length=50, do_sample=False)

print("\n--- LONG TEXT SUMMARY ---")
print(summary[0]['summary_text'])
```

```python
# Save summaries to file

with open("summaries.txt", "w") as f:
    for text in texts:
        summary = summarizer(text, max_length=100, min_length=30, do_sample=False)
        f.write("Original:\n" + text + "\n")
        f.write("Summary:\n" + summary[0]['summary_text'] + "\n\n")

print("Summaries saved to file!")
```
