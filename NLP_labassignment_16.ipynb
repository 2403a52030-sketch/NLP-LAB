```python
import spacy
from spacy import displacy
import pandas as pd

# Load English model
nlp = spacy.load("en_core_web_sm")

print("spaCy model loaded successfully!")
```

```python
# Dataset: Different types of sentences
sentences = [
    "The cat sat on the mat.",
    "She is reading a book.",
    "The boy who is playing is my brother.",
    "I went to the market and bought fruits.",
    "The cake was baked by my mother.",
    "Do you like coffee?",
    "Artificial intelligence is transforming the world.",
    "He quickly finished his homework.",
    "The teacher gave the students homework.",
    "The car that he bought is expensive.",
    "Why are you late?",
    "The match was won by the team.",
    "She sings beautifully.",
    "The dog barked loudly at night.",
    "They are planning a trip to Goa."
]

print("Total Sentences:", len(sentences))
```

```python
docs = [nlp(sentence) for sentence in sentences]

print("Text processed successfully!")
```

```python
all_data = []

for doc in docs:
    for token in doc:
        all_data.append({
            "Sentence": doc.text,
            "Token": token.text,
            "Lemma": token.lemma_,
            "POS": token.pos_,
            "Dependency": token.dep_,
            "Head": token.head.text
        })

# Convert to DataFrame
df = pd.DataFrame(all_data)

# Display first 20 rows
df.head(20)
```

```python
def analyze_sentence(doc):
    subject = ""
    verb = ""
    obj = ""

    for token in doc:
        if token.dep_ == "nsubj":
            subject = token.text
        if token.dep_ == "ROOT":
            verb = token.text
        if token.dep_ in ["dobj", "pobj"]:
            obj = token.text

    return subject, verb, obj


# Analyze all sentences
for doc in docs:
    subj, verb, obj = analyze_sentence(doc)
    print(f"\nSentence: {doc.text}")
    print(f"Subject: {subj}, Verb: {verb}, Object: {obj}")
```

```python
from IPython.display import display, HTML

# Visualize first 5 sentences
for doc in docs[:5]:
    print("\nSentence:", doc.text)
    display(HTML(displacy.render(doc, style="dep", jupyter=True)))
```
